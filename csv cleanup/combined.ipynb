```python
file = "2017.csv"
df2017 = pd.read_csv(file)
df2017.head()
df2017.columns
```

```python
df2017 = df2017.drop(columns=['Whisker.high','Whisker.low'])
df2017.columns
```

```python
df2017_renamed = df2017.rename(columns={'Happiness.Rank': "happiness_rank",
                       'Happiness.Score': "happiness_score",
                       'Economy..GDP.per.Capita.': "gpd_per_capita",
                       'Health..Life.Expectancy.': "life_expectancy",
                       'Trust..Government.Corruption.': "government_corruption",
                       'Dystopia.Residual': "dystopia_residual"
                      })
                                                                                           
```

```python
file = "2018.csv"
df2018 = pd.read_csv(file)
df2018.head()
df2018.columns
```

```python
df2018.insert(loc=1, column="year", value="2018")
df2018
```

```python

df2018_renamed = df2018.rename(columns= {'Overall rank': "happiness_rank", 
                                 'Country or region': "Country",
                                 'Score': "happiness_score",
                                 'GDP per capita': "gpd_per_capita",
                                 'Healthy life expectancy': "life_expectancy",
                                 'Freedom to make life choices': "Freedom",
                                 'Perceptions of corruption': "government_corruption"
                                })
    
```

```python
df2018_renamed.columns
```

```python
frames = [df2015_renamed, df2016_renamed, df2017_renamed, df2018_renamed]

result = pd.concat(frames)
result
```

```python
result = result.drop(columns=['Lower Confidence Interval', 'Upper Confidence Interval'])
```

```python
result = result.fillna("")
result
```

```python
result.to_csv('clean_new.csv', index=False)
```
